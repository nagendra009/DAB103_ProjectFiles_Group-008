## Dataset - Global Terrorism Database by START (Study of Terrorism and Responses to Terrorism)
Link - https://www.start.umd.edu/gtd/access/
#### The GTD is an open-source database, which provides information on domestic and international terrorist attacks around the world since 1970, and now includes more than 200,000 events. For each event, a wide range of information is available, including the date and location of the incident, the weapons used, nature of the target, the number of casualties, and – when identifiable – the group or individual responsible.

## Descriptive Statistics for the dataset

```python
#Filtering dataset to calculate correlation between important columns
gtd = globalterrorismdb[['iyear','imonth','iday','country_txt','region','region_txt','provstate','city'
                         ,'success','suicide','attacktype1_txt','targtype1_txt','weaptype1_txt',
               'nkill','nwound','gname']]
gtd.describe()
```

```python
gtd.describe(include=['O'])
```

## Correlation between columns in the dataframe

```python
gtd_corr = gtd.corr()
round(gtd_corr,2)
```

```python
plt.figure(figsize=(15,10))
hmap = sns.heatmap(gtd_corr);
hmap.set_title("Correlation between Columns in the Dataset", color ="blue", fontsize = 20)
plt.xlabel("Column Names", fontsize=15, color = "darkblue")
plt.ylabel("Column Names", fontsize=15, color = "darkblue")
plt.show()
```

## Analyzing data by each columns

```python
gtd.iyear.value_counts(ascending = False)
```

```python
plt.figure(figsize=(15,10))
plot = sns.countplot(x ='iyear', data = gtd)
plot.set_title("No. of Attacks per Year", fontsize = 20, color ="blue")
plt.xlabel("Years", fontsize=15, color = "darkblue")
plt.ylabel("Number of Attacks", fontsize=15, color = "darkblue")
plot.set_xticklabels(plot.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
gtd.region_txt.value_counts()
```

```python
plt.figure(figsize = (15,10))
plot = sns.countplot(x = 'region_txt', data = gtd)
plot.set_title("Region wise Attacks", fontsize = 20, color = "blue")
plt.xlabel("Regions", fontsize=15, color = "darkblue")
plt.ylabel("Number of Attacks", fontsize=15, color = "darkblue")
plot.set_xticklabels(plot.get_xticklabels(), rotation = 45, horizontalalignment = 'right')
plt.show()
```
